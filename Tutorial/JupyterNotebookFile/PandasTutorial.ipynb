## Getting Specific rows with For loop

```python
for index, row in df.iterrows():
    print(index, row['Attack'])
```

## Finding pokemon with the most Attack Damage

```python
m = 0
pokemon = list()
for index, row in df.iterrows():
    if m <= row['Attack']:
        m = row['Attack']
        pokemon = row

print("Name: ",  pokemon['Name'], "\nAttack: ", pokemon['Attack'])
```
